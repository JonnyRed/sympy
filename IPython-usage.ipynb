## Piecewise function

$$
\begin{align}
    \text{Probability density function:}\\
    \begin{cases}
        \frac{1}{b-a}&\text{for $x\in[a,b]$}\\
        0&\text{otherwise}\\
    \end{cases} \\
    \text{Cumulative distribution function:}\\
    \begin{cases}
        0&\text{for $x<a$}\\
        \frac{x-a}{b-a}&\text{for $x\in[a,b)$}\\
        1&\text{for $x\ge b$}\\
    \end{cases}
\end{align}
$$
